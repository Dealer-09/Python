```python
age = 19
print(type(age))

#Type Conversion
age_str = str(age)
print(type(age_str))

eyes='2' # Character 
print(type(int(eyes)))
```

```python
#Simple Calculator
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

sum = num1 + num2
difference = num1 - num2
product = num1 * num2
quotient = num1 / num2

print("Sum:", sum)
print("Difference:", difference)
print("Product:", product)
print("Quotient:", quotient)
```
