```python
## Create a set 
my_set = {1, 2, 3, 4, 5}
print(my_set)   
print(type(my_set))
```

```python
my_empty_set=set([1,2,3,4,4,4,5,5,6])  ## [1,2,3,4,4,4,5,5,6]
print(my_empty_set)
```

## Basic set operations

```python
# Adding & Removal of elements
my_set.add(6)
print(my_set)
my_set.add(6)
print(my_set) ##Repeated elements are ignored
```

```python
## Removal of of elements
my_set.remove(3)
print(my_set)
```

```python
# my_set.remove(7)   ## No element 7 in the set, so it will throw an error
```

```python
my_set.discard(7) ## No element 7 in the set, but it will not throw an error
```

```python
## Pop method
removed_element = my_set.pop()  ## It will remove and return an arbitrary element from the set
print(removed_element)
print(my_set)
```

```python
my_set.clear()  ## It will remove all elements from the set
print(my_set)
```

```python
## Set Membership Test
my_set={1, 2, 3, 4, 5}
print(3 in my_set)  ## True
print(6 in my_set)  ## False
```

## Mathematics set operations

```python
set1 = {1, 2, 3, 4, 5, 6}
set2 = {4, 5, 6, 7, 8, 9}

## Union
union_set = set1.union(set2)
print(union_set)

## Intersection
intersection_set = set1.intersection(set2)
print(intersection_set)
set1.intersection_update(set2)  ## It will update set1 with the intersection of set1 and set2
print(set1)

```

```python
set1 = {1, 2, 3, 4, 5, 6}
set2 = {4, 5, 6, 7, 8, 9}

## Difference
difference_set = set1.difference(set2)
print(difference_set)
```
